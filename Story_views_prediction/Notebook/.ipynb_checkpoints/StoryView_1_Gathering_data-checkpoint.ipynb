# Formula Crating

```python
# Utility
def get_units(n):
    temp_n =n
    unit=0
    while temp_n > 9:
        temp_n = round(temp_n/10)
        unit= unit+1
    return unit

def roundoff(number,digit=1):
    temp_num = number
    temp_num = temp_num/(10**digit)
    temp_num = round(temp_num)
    return temp_num*(10**digit)

def range_it(nm):
    low_price = nm*0.9
    high_price = nm*1.1
    low_price = roundoff(low_price,get_units(low_price)-1)
    high_price = roundoff(high_price,get_units(high_price)-1)
    return low_price,high_price
```

```python
def story_view_1(followers,eng_rate=2):
    eng_num = followers*eng_rate/100
    positive_effect = (1+7.781641027922785)*eng_num
    neutral_effect = (1+0.4126992022969042)*eng_num
    negative_effect = (1-0.42657390782570154)*eng_num
    return roundoff(positive_effect,get_units(positive_effect)-2),roundoff(neutral_effect,get_units(neutral_effect)-2),roundoff(negative_effect,get_units(negative_effect)-2)

# 4.5393066246473967 | 3.0219542397250615 | nil
```

```python
def story_view_2(followers,eng_rate=2):
    eng_num = followers*eng_rate/100
    reach = 0.01524404*followers + 1.20821245*eng_num
    return roundoff(reach,get_units(reach)-2)
```

## Transient and used case testing

```python
used_case_df = pd.DataFrame(columns = ["followers","eng_rate","Actual_reach","F1_positive","F1_neutral","F1_negative","F2_reach"])
display(story_view_df)
for i in story_view_df.index:
    pe,nue,nee = story_view_1(story_view_df.followers.loc[i],story_view_df.avg_engagement.loc[i])
    f2_reach = story_view_2(story_view_df.followers.loc[i],story_view_df.avg_engagement.loc[i])
    push_dict={"followers":story_view_df.followers.loc[i],
               "eng_rate":story_view_df.avg_engagement.loc[i],
               "Actual_reach":story_view_df.reach.loc[i],
               "F1_positive":pe,
               "F1_neutral":nue,
               "F1_negative":nee,
               "F2_reach":f2_reach}
    used_case_df = used_case_df.append(push_dict,ignore_index=True)
used_case_df = used_case_df.sort_values(by='followers')
display(used_case_df)
used_case_df.to_csv(r'E:\Winkl Mains\Task_3_StoryView_prediction\DATA\story_used_cases.csv')
```
