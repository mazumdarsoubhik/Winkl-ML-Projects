# Database Influencer Analysis 

```python
insta_db = pd.read_csv('Database_influencer_price.csv')
insta_db = insta_db.dropna()
insta_nano = insta_db[(insta_db['followers'] >1000) & (insta_db['followers'] <10000)]
insta_micro = insta_db[(insta_db['followers'] >10000) & (insta_db['followers'] <200000)]
insta_macro = insta_db[(insta_db['followers'] >200000) & (insta_db['followers'] <1000000)]
insta_celeb = insta_db[insta_db['followers'] >1000000]
print(len(insta_nano),len(insta_micro),len(insta_macro),len(insta_celeb))
insta_db.head()
```

## Influencer image charge analysis

```python
insta_image_nano = insta_nano[insta_nano.image !=0]
insta_image_micro = insta_micro[insta_micro.image !=0]
insta_image_macro = insta_macro[insta_macro.image !=0]
insta_image_celeb = insta_celeb[insta_celeb.image !=0]

def compare_with_predicted(data_frame,story,image,video,reel,carousel):
    insta_pred = pd.DataFrame(columns = ["followers","avg_eng","given_amount","pred_amount"])
    for i in data_frame.index:
        push_dict = {"followers":data_frame.followers.loc[i],
                     "avg_eng":data_frame.avg_engagement.loc[i],
                     "given_amount":data_frame.image.loc[i],
                     "pred_amount":new_price_2(data_frame.followers.loc[i],data_frame.avg_engagement.loc[i],story,image,video,reel,carousel)        
        }
        insta_pred = insta_pred.append(push_dict,ignore_index=True)
    insta_pred = insta_pred.sort_values("followers")
    insta_pred['rolling_avg'] = insta_pred.given_amount.rolling(50).mean()
    return insta_pred

insta_nano_pred = compare_with_predicted(insta_image_nano,0,1,0,0,0)
insta_micro_pred = compare_with_predicted(insta_image_micro,0,1,0,0,0)
insta_macro_pred = compare_with_predicted(insta_image_macro,0,1,0,0,0)
insta_celeb_pred = compare_with_predicted(insta_image_celeb,0,1,0,0,0)

def plot_difference(data_frame):
    plt.figure(figsize=(15, 10), dpi=100)
    plt.plot(data_frame['followers'].tolist(), data_frame['given_amount'].tolist(),label="Influencer Amount")
    plt.plot(data_frame['followers'].tolist(), data_frame['pred_amount'].tolist(),label="Predicted Amount")
    plt.plot(data_frame['followers'].tolist(), data_frame['rolling_avg'].tolist(),label="Average Amount")
        
    plt.xlabel('Followers unit->10M')
    # Set the y axis label of the current axis.
    plt.ylabel('Price')
    # Set a title of the current axes.
    plt.title('Instagram Transient Pricing Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_difference(insta_nano_pred)
plot_difference(insta_micro_pred)
plot_difference(insta_macro_pred)
plot_difference(insta_celeb_pred)

# plt.plot(insta_video_pred['followers'].tolist(), insta_video_pred['given_amount'].tolist(),label="Influencer Amount")
# # plt.plot(insta_video_pred['followers'].tolist(), insta_video_pred['pred_amount'].tolist(),label="Predicted Amount")
# plt.xlabel('Followers unit->10M')
# # Set the y axis label of the current axis.
# plt.ylabel('Price')
# # Set a title of the current axes.
# plt.title('Instagram Transient Pricing Analysis')
# plt.legend()
# plt.grid(True)
# plt.show()
```
