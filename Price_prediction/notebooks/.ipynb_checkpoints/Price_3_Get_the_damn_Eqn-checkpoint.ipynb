```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score

# Feature importance
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

# Insta Image

```python
insta_image_df = pd.read_csv(r'E:\Winkl Mains\Task_2 Price Prediction\DATA\Price_insta_image.csv')
insta_image_df = insta_image_df[insta_image_df.amount != 0]
display(insta_image_df.tail())
```

```python
# Mandatory Second Run only
for i in insta_image_df.index:
    insta_image_df.amount.loc[i] = insta_image_df.amount.loc[i]/insta_image_df.insta_image.loc[i]
    insta_image_df.insta_image.loc[i] =1
display(insta_image_df.tail())
```

```python
insta_image_follow = insta_image_df[["followers","amount"]]
display(insta_image_follow.head())
```

```python
X = insta_image_follow['followers'].values[:,np.newaxis]
y = insta_image_follow['amount'].values

model2 = LinearRegression(fit_intercept=False)
model2.fit(X, y)

plt.scatter(X, y,color='g')
plt.plot(X, model2.predict(X),color='k')

plt.show()

#Manual Check
influ_followers = 10000
print("Predicted: ",model2.predict([[influ_followers]]))

# Coefficients
print("coefficient: ",model2.coef_)
print("intercept: ",model2.intercept_)
```

## Creating Formula

```python
def price_1(followers,story,image,video,reel):
    story_rate = 0.0171571*story
    image_rate = 0.02512316*image
    video_rate = 0.07270367*video
    reel_rate = 0.15138649*reel
    total_price = (story_rate+image_rate+video_rate+reel_rate)*followers
    return total_price
```

```python
print(price_1(71608,1,0,1,0))
```

## Testing Formula

```python
insta_data = pd.read_csv(r'E:\Winkl Mains\Task_2 Price Prediction\DATA\InstagramData.csv')
insta_data = insta_data[["followers","insta_image","insta_story","insta_video","insta_reel","amount"]]
insta_data.head()
insta_pred = pd.DataFrame(columns = ["followers","insta_image","insta_story","insta_video","insta_reel","amount","pred_amount"])
for i in insta_data.index:
    push_dict= {"followers":insta_data.followers.loc[i],
                "insta_image":insta_data.insta_image.loc[i],
                "insta_story":insta_data.insta_story.loc[i],
                "insta_video":insta_data.insta_video.loc[i],
                "insta_reel":insta_data.insta_reel.loc[i],
                "amount":insta_data.amount.loc[i],
                "pred_amount":price_1(insta_data.followers.loc[i],insta_data.insta_story.loc[i],insta_data.insta_image.loc[i],insta_data.insta_video.loc[i],insta_data.insta_reel.loc[i])
    }
    insta_pred = insta_pred.append(push_dict,ignore_index=True)
display(insta_pred)
insta_pred.to_csv(r'E:\Winkl Mains\Task_2 Price Prediction\DATA\Insta_pred_prices.csv')
```

## Formula improvement and Testing

```python
def price_2(followers,story,image,video,reel):
    story_rate = 0.0171571*story
    image_rate = 0.02512316*image
    video_rate = 0.07270367*video
    reel_rate = 0.15138649*reel
    total_price = (story_rate+image_rate+video_rate+reel_rate)*followers
    return (total_price*1.08562297961)

insta_data = pd.read_csv(r'E:\Winkl Mains\Task_2 Price Prediction\DATA\InstagramData.csv')
insta_data = insta_data[["followers","insta_image","insta_story","insta_video","insta_reel","amount"]]
insta_data.head()
insta_pred2 = pd.DataFrame(columns = ["followers","insta_image","insta_story","insta_video","insta_reel","amount","pred_amount"])
for i in insta_data.index:
    push_dict= {"followers":insta_data.followers.loc[i],
                "insta_image":insta_data.insta_image.loc[i],
                "insta_story":insta_data.insta_story.loc[i],
                "insta_video":insta_data.insta_video.loc[i],
                "insta_reel":insta_data.insta_reel.loc[i],
                "amount":insta_data.amount.loc[i],
                "pred_amount":price_2(insta_data.followers.loc[i],insta_data.insta_story.loc[i],insta_data.insta_image.loc[i],insta_data.insta_video.loc[i],insta_data.insta_reel.loc[i])
    }
    insta_pred2 = insta_pred2.append(push_dict,ignore_index=True)
display(insta_pred2)
insta_pred2.to_csv(r'E:\Winkl Mains\Task_2 Price Prediction\DATA\Insta_pred_prices2.csv')
```
